```python
def hieght(X,w):
  Xw=np.dot(X,w)
  return Xw
```

```python
def love(Xw,y,X):
  y= np.array([[1],[2],[3],[4],[5]])
  m=X.shape[0]
  jx=(1/2*m)*np.sum(Xw - y)**2
  return jx
```

```python
def denta(Xw,X,y):
  m=X.shape[0]
  denta_jw=(np.sum(np.dot(X.T,(Xw-y))))/m
  return denta_jw
```

```python
def lucky(X, y, w, alpha):
  W = w- alpha*denta(Xw,X,y)
  return W
```

```python
def main():
  ex=hieght(X,w)
  print("vector hw(X)",ex)
  happy=love(Xw,y,X)
  print("Jx la:",happy)
  smile=denta(Xw,X,y)
  print("denta Jx",smile)

  alpha= 0.5
  luc=lucky(X, y, w, alpha)
  print("w la",luc)
if __name__ == "__main__":
  main()
```
